# 2. All you need is plot

The most popular way to graph data in Python is with the Matplotlib library. If you have values for one variable in an array ``x`` and the values for another variable in an array ``y``, you can create a graph of ``y`` versus ``x`` with a simple plot using Matplotlib's ``plot`` command, like so:

````python
import numpy as np # Import numpy library.
import matplotlib.pyplot as plt # Import Matplotlib's pyplot library.
# Show plotting results in the output.
%matplotlib inline 

# Generate data arrays.
x = np.linspace(-4,4,100)
y = np.exp(-x**2)

# Create a figure.
MyFigure, MyPlot = plt.subplots()

# Plot data.
MyPlot.plot(x,y)
````

# 5. Sometimes you figure out data as you go: Error in $e$

The above examples assumed that we could get all our data values at once, but this isn't always possible. In many physics applications, **we compute data values as we go** and need to add them one at a time to our array. For example, let's suppose we wanted to graph the **current estimate** for $e$ as we add progressively more terms in the series expansion from Python Tutorials 1 & 2. We would need to add the current estimate one at a time to the array.

In order to do this, we need to separate the process of **creating an array** from the process of **filling an array**. Numpy offers a handy way to do this with its ``np.empty()`` function. It literally creates an empty array that we can fill in later!

The code cell below carries out this process:

````python
import numpy as np 
import matplotlib.pyplot as plt
% matplotlib inline

# Set up an array to count the terms we're adding.
TermCount = np.array( [1,2,3,4] )

# Set up an empty array to store our estimates of e.
eEstimates = np.empty( 4 ) # What's significant about this value of 4?

Terms = np.array( [ 1, 1, 1/(2*1), 1/(3*2*1) ] )

eEstimate = Terms[0] 

eEstimates[0] = eEstimate

eEstimate = eEstimate + Terms[1] 

eEstimates[1] = eEstimate

eEstimate = eEstimate + Terms[2] 

eEstimates[2] = eEstimate

eEstimate = eEstimate + Terms[3] 

eEstimates[3] = eEstimate

# Now graph the e estimates versus the term count.
MyFigure,MyPlot = plt.subplots()
MyPlot.plot(TermCount,eEstimates)
````
