# 2. All you need is plot

If you have values for one variable in an array ``x`` and the values for another variable in an array ``y``, you can create a graph of ``y`` versus ``x`` with a simple plot using Matlab's ``plot`` command, like so:

````matlab
% Generate data arrays.
x = linspace(-4,4,100);
y = exp(-x.^2);

% Plot data.
plot(x,y)
````

# 5. Sometimes you figure out data as you go: Error in $e$

The above examples assumed that we could get all our data values at once, but this isn't always possible. In many physics applications, **we compute data values as we go** and need to add them one at a time to our array. For example, let's suppose we wanted to graph the **current estimate** for $e$ as we add progressively more terms in the series expansion from Matlab Tutorials 1 & 2. We would need to add the current estimate one at a time to the array.

In order to do this, we can take advantage of how **Matlab will expand the size of an array as we add elements to it**.

The code cell below carries out this process:

````matlab
% Set up an array to count the terms we're adding.
TermCount = [1,2,3,4]

Terms = [ 1, 1, 1/(2*1), 1/(3*2*1) ] 

% Set up an array to store our estimates of e.
eEstimates = Terms(1)

eEstimates(2) = Terms(1) + Terms(2) 

eEstimates(3) = Terms(1) + Terms(2) + Terms(3)

eEstimates(4) = Terms(1) + Terms(2) + Terms(3) + Terms(4)

% Now graph the e estimates versus the term count.
plot(TermCount,eEstimates)
````
